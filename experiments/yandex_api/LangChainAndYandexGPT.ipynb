# Import libs

```python
import os

from langchain_core.messages import HumanMessage, SystemMessage
from langchain_openai import ChatOpenAI
```

# Define API

```python
YC_API_KEY = os.getenv("API_KEY")
FOLDER_ID = os.getenv("FOLDER_ID")
model_name = f"gpt://{FOLDER_ID}/gpt-oss-120b/latest"
```

```python
llm = ChatOpenAI(
    model=model_name,
    api_key=YC_API_KEY,
    base_url="https://llm.api.cloud.yandex.net/v1",
    temperature=0.7,
)
```

```python
llm
```

```python
# Простой запрос
messages = [
    SystemMessage(
        content="Ты опытный аналитик по цифровому развитию который помогает младшему сотруднику подготовиться к конференции"
    ),
    HumanMessage(
        content="Помоги составить текст с текущими трендами 2025 года по созданию AI агентов в банковском секторе."
    ),
]
```

```python
response = llm.invoke(messages)
print(response.content)
```
